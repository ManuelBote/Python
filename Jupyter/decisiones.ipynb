# Predicciones sobre datos no numericos: Arbol de decisiones

```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier as dtc
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('comediantes.csv')

print(df)
```

```python
d = {'UK':0, 'USA':1, 'N':2}
df['Nationality'] = df['Nationality'].map(d)

d = {'NO':0, 'YES':1}
df['Go'] = df['Go'].map(d)

print(df)
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']
X = df[features]
y = df['Go']

print(X)
print(y)
```

```python
dtree = dtc()

dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names = features)
```
