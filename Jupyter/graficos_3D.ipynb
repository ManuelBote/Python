# Representacion en 3 dimensiones con Python utilizando Matplotlib

Los graficos tridimensionales son herramientas muy importantes para visualizar datos que constan de 3 dimensiones, tales como datos que presentan dos variables dependientes y una independiente. Representando graficamente los datos en diagramas 3D podemos tener un mejor entendimiento de los datos que poseen 3 variables. Podremos usar varias funciones de la libreria Matplotlib para conseguir nuestro proposito.

## Ejemplo de gradico tridimensional empleando Matplotlib

Comenzaremos primero representando los ejes tridimensionales usando la libreria Matplotlib. Para lograrlo, simplemente tenemos que cambiar el parametrpo de proyeccion de la funcion ```plt.axes()```

````python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection = '3d')
````

Con los ejes tridimensionales ya proyectados podemos ahora ser capaces de representar graficos en 3 dimensiones. Dichos graficos nos dan una aproximacion mas dinamica y hacen que los datos sean mas interactivo. A l igual que sucedia con el grafico bidimensional, podemos usar diferentes formas para realizar graficos 3D (por ejemplo, graficos de dispersion, graficos de entorno, graficos de superficies, etc)

Los graficos mas simples son aquellos que consisten en lineas y puntos, los cuales realizaremos con las funciones ```ax.plot3D()``` y ```ax.scatter()```

## Graficos de lineas tridimensionales empleando Matplotlib

Para representar grficamente diagramas de lineas usaremos la funcion `plot3D()` de la libreria mpl_toolkits. Para trazar lineas en 3D tendremos que inicializar 3 variables para la ecuacion de linea. En nuestro caso, definiremos 3 variables como _x_,_y_,_z_

````python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection = '3d')

z = np.linspace(0,1,100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)

ax.plot3D(x,y,z, "green")
ax.set_title("Grafico de linea tridimensional de una funcion sinucoidal")
plt.show()
````

## Grafico de dispersion tridimensional empleando Matplotlib

Para representar graficamente el mismo conjunto de datos anterior mediante un diagrama de dispersion, usaremos la funcion `scatter()` de la libreria mpl_toolkits. El resultado sera el mismo que anteriormente pero a traves de puntos bien diferenciados

````python
fig = plt.figure()
ax = plt.axes(projection = '3d')

z = np.linspace(0,1,100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)
color = x + y

ax.scatter(x,y,z,c = color)
ax.set_title("Grafico de dispersion tridimensional de una funcion sinuidal")

plt.show()
````

## Grafico de superficie empleando Matplotlib

Los graficos de superficies y de alambre se encargan de trabajar sobre datos cuadriculados. Dichos graficos toman el valor cuadriculado y lo representan en una superficie tridimensional. Para poder realizar este representacion, usaremos la funcion `plot_surface()`, la cual nos permite trazar la superficie deseada.

````python
x = np.outer(np.linspace(-2,2,10), np.ones(10))
y = x.copy().T
z = np.cos(x**2 + y**3)

fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_surface(x, y, z, cmap = 'viridis', edgecolor = 'green')

plt.show()
````

## Grafico de alambre empleando Matplotlib

Para dibujar un grafico de alambre utilizaremos la funcion `plot_wireframe()` perteneciente a la libreria mpl_toolkits

````python
def f (x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-1,5,10)
y = np.linspace(-1,5,10)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_wireframe(X,Y,Z,color="green")

plt.show()
````

## Graficos de contorno empleando Matplotlib

Los grficos de contorno toman los datos de entrada y los disponen en redes bidimensionales. Despus, los datos del eje Z son evaluados en cada punto. Para poder trazar un grfico de contorno emplearemos de forma combinada las funciones `meshgrid` y `plot_surface()`. Esta clase de grficos resulta una excelente manera de visualizar grficos de optimizacin.

````python
def function(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-10,10,40)
y = np.linspace(-10,10,40)

X, Y = np.meshgrid(x,y)
Z = function(X,Y)

fig = plt.figure(figsize = (10, 8))
ax = plt.axes(projection = '3d')

ax.plot_surface(X, Y, Z, cmap = 'cool', alpha = 0.8)
ax.set_xlabel('x', fontsize = 12)
ax.set_ylabel('y', fontsize = 12)
ax.set_zlabel('z', fontsize = 12)

plt.show()
````

## Ejemplo practico: representacion gradfica de la banda de Mobius

#### Que es la banda de Mobius?

Se trata de un "objeto" matematico (tambien llamado cilindro torcido) que tiene la peculiaridad de ser una superficie de una unica cara y sin bordes. Matematicamente hablando, este hecho se refleja en que la banda de Mobius es una superficie no orientable. Para poder representarla graficamente, necesitamos parametrizar la superficie en la cual se apoyara la banda (tal y como hemos hecho en ejemplos anteriores).

````python
R = 2

u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u, v)

x = (R + v * np.cos(u/2)) * np.cos(u)
y = (R + v * np.cos(u/2)) * np.sin(u)
z = v * np.sin(u/2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

plt.show()
````

## Grafico tridimensional con el angulo de vision modificado

A continuacion , veremos como podemos variar el angulo sobre el cual visualizamos nuestros graficos en tres dimensiones. Para poder observar nuestras figuras desde cualquier punto, deberemos emplear la funcion `view_init()` de la libreria mpl_toolkit. Dicha funcion, como su nombre indica, establece la vista inicial de nuestro grafico representado.

````python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection = '3d')

z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z, "green")
ax.view_init(160, 30)
plt.show()
````

## Grafico tridimensional animado empleando Matplotlib

Las representaciones grficas siempre son sencillas de entender y observar, de cara a una mejor comunicacin. Adems de modificar el ngulo de visin de nuestros grficos, mediante Matplotlib podemos animar dichas figuras, de modo que la visualizacin de estas adquiera un carcter ms interactivo an. Para conseguir esto, simplemente debemos de ir variando el ngulo de visin en cada instante de tiempo, lo cual se logra iterando (ediante una estructura de repeticin de Python) la funcin `view_init()` para cada valor de ngulo de visin.

````python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection = '3d')

z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z, "green")

for angle in range(0, 360):
    ax.view_init(angle, 30)
    plt.draw()
    plt.pause(0.001)

plt.show()
````
