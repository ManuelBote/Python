# Manipulacionde datos representados: Regresion Multiple

```python
import pandas as pd

df = pd.read_csv("data.csv")

print(df)
```

```python
X = df[['Weight', 'Volume']]
y = df['CO2']

print("Valores de la variable independiente:")
print(X)
print("Valores de la variable dependiente:")
print(y)
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
predictedCO2 = regr.predict([[2300, 1300]])
print("Las emisiones de un coche de 2300 kilos y 1300 centimetros cubicos de motor, son:", predictedCO2)
```

## Verificacion de la prediccion respecto a las variedades independientes: Coeficiente de Regresion

```python
predictedCO2=regr.predict([[3300,1300]])
print('Las emisiones de CO2 de un coche 1000 veces más pesado son: ',predictedCO2)
```
