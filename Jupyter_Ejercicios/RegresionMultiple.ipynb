# Regresion Multiple entre varias variables

## Peso de una persona, pulso card√≠aco de una persona y colesterol de una persona.

```python
import pandas as pd

df = pd.read_csv("persona.csv")

print(df)
```

```python
X = df[['Peso', 'Pulso']]
y = df['Colesterol']

print("Valores de la variable independiente:")
print(X)
print("Valores de la variable dependiente:")
print(y)
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
predictedColesterol = regr.predict([[85, 100]])
print("El colesterol de una persona de 85kg de peso y 100 pulsaciones/min es:", predictedColesterol)
```
