# REGRESIÓN LINEAL ENTRE DOS VARIABLES

## Número de libros leídos por una persona y número de pasos dados por una persona (al día).

```python
import matplotlib.pyplot as plt
from scipy import stats

pasos = [4000, 5300, 3100, 12600, 7100, 4900, 6400, 2000, 9800, 11600, 5400, 3200, 7500, 5500, 4200, 
         6800, 7000, 5500, 9500, 11000, 8000, 11500, 5000, 8900, 6000, 12000, 13000, 8000, 8500, 7000]
libros = [10, 8, 13, 9, 11, 14, 6, 7, 5, 3, 4, 12, 16, 15, 10, 18, 20, 25, 22, 17, 8, 14, 9, 7, 11, 13, 6, 18, 10, 12]


slope, intercept, r, p, std_err = stats.linregress(pasos, libros)

def myfunc (x):
    return slope * x + intercept

mymodel = list(map(myfunc, pasos))

plt.scatter(pasos, libros)
plt.xlabel("Pasos")
plt.ylabel("Libros")
plt.plot(pasos, mymodel)
plt.show()

print("El coeficiente de correlacion entre ambas variables es:", r)
```

En este caso, al se puede saber que ambos datos no tienen ninguna relacion entre ellos, ya que el coeficiente de correlacion es muy cercano a 0.

## Altura de una persona y peso de una persona.

```python
altura = [160, 162, 165, 170, 172, 175, 178, 180, 185, 188, 190, 192, 194, 198, 200, 163, 167, 172, 174, 177, 180, 185, 168, 171, 173, 177, 182, 184, 189, 191]
peso = [55, 57, 60, 65, 67, 70, 72, 75, 80, 85, 88, 90, 92, 95, 97, 56, 59, 64, 66, 70, 74, 79, 60, 63, 66, 70, 76, 78, 83, 86]


slope, intercept, r, p, std_err = stats.linregress(altura, peso)

def myfunc (x):
    return slope * x + intercept

mymodel = list(map(myfunc, altura))

plt.scatter(altura, peso)
plt.xlabel("Altura")
plt.ylabel("Peso")
plt.plot(altura, mymodel)
plt.show()

print("El coeficiente de correlacion entre ambas variables es:", r)
```

En este caso el coeficiente de correlacion esta muy cerca del 1, por lo que se puede afirmar que ambos datos estan relacionados.
